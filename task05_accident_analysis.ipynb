```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("global_traffic_accidents.csv")

# Convert Date + Time to datetime
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])

# Extract Hour & Month
df['Hour'] = df['DateTime'].dt.hour
df['Month'] = df['DateTime'].dt.month

# 1. Plot accidents by Hour
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Hour', palette='coolwarm')
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# 2. Top 10 Weather Conditions
plt.figure(figsize=(10,8))
sns.countplot(data=df, y='Weather Condition', order=df['Weather Condition'].value_counts().head(10).index, palette='magma')
plt.title("Top 10 Weather Conditions in Accidents")
plt.xlabel("Count")
plt.ylabel("Weather Condition")
plt.tight_layout()
plt.show()

# 3. Top 10 Road Conditions
plt.figure(figsize=(10,8))
sns.countplot(data=df, y='Road Condition', order=df['Road Condition'].value_counts().head(10).index, palette='Blues')
plt.title("Top 10 Road Conditions in Accidents")
plt.xlabel("Count")
plt.ylabel("Road Condition")
plt.tight_layout()
plt.show()

# 4. Severity proxy â€“ Casualties distribution
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Casualties', bins=20, kde=True, color='red')
plt.title("Distribution of Casualties in Accidents")
plt.xlabel("Number of Casualties")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# 5. Vehicles involved
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Vehicles Involved', bins=10, kde=True, color='green')
plt.title("Number of Vehicles Involved in Accidents")
plt.xlabel("Vehicles Involved")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
